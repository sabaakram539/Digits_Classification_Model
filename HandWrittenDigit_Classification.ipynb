```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.datasets import load_digits
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target,train_size=0.8,random_state=42)
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

In Random Forest Classifier we saw that the Accuracy Score was as high as 97.22%, very high as compared to the Decision Tree model.
